Расчёт прибыли:
--
$$\overline{x_{м1}} - \overline{x_д} = z_1$$
$$( x_{p1} = \frac{z_1}{x_{д1}} )$$ 
$$\overline{x_{м2}} - \overline{x_д} = z_2$$
$$( x_{p2} = \frac{z_2}{x_{д2}} )$$ 

,где x_д - среднее значение события, x_м1 - первое макс. значение, x_м2 - второе макс. значение, z1 - результат вычитания из x_м1/абсолютное значение прибыли относительно x_м1, z2 - результат вычитания из x_м2/абсолютное значение прибыли относительно x_м2, x_p1 - относительное(%) значение прибыли относительно x_м1, x_p2 - относительное(%) значение прибыли относительно x_м2.

USD: 
--

111,76 - 75 = 36,76 (𝑥_𝑝1 ~ 49 %)

120,38 - 75 = 45,38  (𝑥_𝑝2 ~ 60,5 %)


СNY: 
--
17,69 - 11,9 = 5,79 (𝑥_𝑝1 ~ 48,66 %)

19 - 11,9 = 7,1 (𝑥_𝑝2 ~ 59,66 %)


СHF: 
--
121,45 - 82 = 39,45 (𝑥_𝑝1 ~ 48,1 %)

129,84 - 82 = 47,84 (𝑥_𝑝2 ~ 58,34 %)

Стоит держать в глове что это максимальная (невозможная) прибль, что будут уплочены 2 комиссии (при покупке и продаже), что биржи будут закрыты в самый благоприятный момент для спекуляции и что банки и прочии конторы драли большую комиссию (Сбер при курсе бакса ЦБ - 120 покупал за 100)

Расчёт прибыли:
--
$$\overline{x_{м1}} - \overline{x_д} = z_1$$
$$( x_{p1} = \frac{z_1}{x_{д1}} )$$ 
$$\overline{x_{м2}} - \overline{x_д} = z_2$$
$$( x_{p2} = \frac{z_2}{x_{д2}} )$$ 

,где x_д - среднее значение события, x_м1 - первое макс. значение, x_м2 - второе макс. значение, z1 - результат вычитания из x_м1/абсолютное значение прибыли относительно x_м1, z2 - результат вычитания из x_м2/абсолютное значение прибыли относительно x_м2, x_p1 - относительное(%) значение прибыли относительно x_м1, x_p2 - относительное(%) значение прибыли относительно x_м2.

Расчёт прибыли:
--
$$\overline{x_{м1}} - \overline{x_д} = z_1$$
$$( x_{p1} = \frac{z_1}{x_{д1}} )$$ 
$$\overline{x_{м2}} - \overline{x_д} = z_2$$
$$( x_{p2} = \frac{z_2}{x_{д2}} )$$ 

,где x_д - среднее значение события, x_м1 - первое макс. значение, x_м2 - второе макс. значение, z1 - результат вычитания из x_м1/абсолютное значение прибыли относительно x_м1, z2 - результат вычитания из x_м2/абсолютное значение прибыли относительно x_м2, x_p1 - относительное(%) значение прибыли относительно x_м1, x_p2 - относительное(%) значение прибыли относительно x_м2.
